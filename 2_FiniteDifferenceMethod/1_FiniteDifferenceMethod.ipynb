# Chapter1 有限差分法

p36の図2.18

厳密解の増幅率，位相誤差は

$$
g = \exp (- \mu \theta^2)
$$

```python
def g_exact(theta, mu):
	return np.exp(- mu * theta**2)

def g_scheme(theta, mu):
	return 1 - 2 * mu * (1 - np.cos(theta))

x = np.linspace(0, np.pi, 100)


fig, axes = plt.subplots(nrows=6, ncols=1, facecolor="white", figsize=(8, 20), sharex=True)
for i, mu in enumerate([-0.1, 0, 0.25, 0.5, 0.75, 1.0]):
	ax = axes[i]
	y_exact = g_exact(theta=x, mu=mu)
	y_scheme = g_scheme(theta=x, mu=mu)
	ax.plot(x, y_exact, label=r"$\mu={mu}$ exact".format(mu=mu))
	ax.plot(x, y_scheme, label=r"$\mu={mu}$ scheme".format(mu=mu))
	ax.set_ylim([-1.1, 1.1])

	ax.set_ylabel(r"$g$")
	ax.legend(fontsize=10)
ax.set_xlabel(r"$\theta$")
```
